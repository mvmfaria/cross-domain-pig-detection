```python
import matplotlib.pyplot as plt
model_data = [
    {"name": "yolov8n", "params": 3.2, "size": "n", "mAP": 0.880, "family": "YOLOv8"},
    {"name": "yolov8s", "params": 11.2, "size": "s", "mAP": 0.906, "family": "YOLOv8"},
    {"name": "yolov8m", "params": 25.9, "size": "m", "mAP": 0.918, "family": "YOLOv8"},
    {"name": "yolov8l", "params": 43.7, "size": "l", "mAP": 0.919, "family": "YOLOv8"},
    {"name": "yolov8x", "params": 68.2, "size": "x", "mAP": 0.919, "family": "YOLOv8"},
    {"name": "yolo12n", "params": 2.6, "size": "n", "mAP": 0.878, "family": "YOLO12"},
    {"name": "yolo12s", "params": 9.3, "size": "s", "mAP": 0.902, "family": "YOLO12"},
    {"name": "yolo12m", "params": 20.2, "size": "m", "mAP": 0.911, "family": "YOLO12"},
    {"name": "yolo12l", "params": 26.4, "size": "l", "mAP": 0.912, "family": "YOLO12"},
    {"name": "yolo12x", "params": 59.1, "size": "x", "mAP": 0.913, "family": "YOLO12"}
]
families = ["YOLOv8", "YOLO12"]
colors = {"YOLOv8": "blue", "YOLO12": "red"}
alphas = {"YOLOv8": 1.0, "YOLO12": 0.3}
plt.figure(figsize=(6,4))
for fam in families:
    fam_models = [m for m in model_data if m["family"] == fam]
    x = [m["params"] for m in fam_models]
    y = [m["mAP"]*100 for m in fam_models]
    labels = [m["size"] for m in fam_models]
    plt.plot(x, y, marker='o', color=colors[fam], label=fam, alpha=alphas[fam])
    for xi, yi, label in zip(x, y, labels):
        plt.text(xi, yi+0.1, label, fontsize=12, color=colors[fam], alpha=alphas[fam], ha='center', va='bottom')
plt.xlabel("Parameters (M)")
plt.ylabel("COCO mAP (%)")
plt.legend(loc='lower right')
plt.grid(True, alpha=0.3)
plt.show()
```
